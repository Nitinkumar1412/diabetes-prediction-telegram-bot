```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, classification_report
```

```python
data = pd.read_csv("C:\\Users\\Nitin\\OneDrive\\Desktop\\BTECH 2nd sem\\Diabetes_prediction_AgenticAi\\Dataset\\pima_diabetes_preprocessed.csv")
data.head()
```

```python
x=data.drop(["Outcome","DiabetesPedigreeFunction"],axis=1)
y=data["Outcome"]
#test train split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
scalar = StandardScaler()
X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)

model = RandomForestClassifier(n_estimators=600,random_state=42,class_weight="balanced")
model.fit(X_train_scaled,y_train)
x.head()
```

```python
y_pred = model.predict(X_test_scaled)
print("accuracy:",accuracy_score(y_test,y_pred))
print("confusion matrix: ",confusion_matrix(y_test,y_pred))
print("\nclassification_report \n",classification_report(y_test,y_pred))
```
